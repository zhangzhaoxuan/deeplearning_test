```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
%matplotlib inline
```

```python
datasets = pd.read_csv("heart.csv")
datasets.head()
```

```python
datasets.shape
```

```python
datasets.info()
```

```python
datasets.describe()
```

```python
datasets.corr()  #各个特征之间的相关性系数
```

```python
from sklearn.model_selection import train_test_split #划分训练集和验证集
train_set, test_set = train_test_split(datasets, test_size=0.2)
```

```python
train_set #显示效果
```

```python
train_set = np.array(train_set)  #将DataFrame转换为ndarray
test_set = np.array(test_set)
x_train, y_train, x_test, y_test = train_set[:, :13], train_set[:, 13:], test_set[:, :13], test_set[:, 13:]
x_train.shape, y_train.shape #验证
y_train = y_train.ravel()  #扁平化,然后进行拟合
```

## 随机森林分类器

```python
from sklearn.ensemble import RandomForestClassifier  #简单的随机森林分类器
rfc_tree = RandomForestClassifier()
rfc_tree.fit(x_train, y_train)
```

```python
y_test_pred = rfc_tree.predict(x_test)
```

```python
from sklearn.metrics import precision_score  #评估准确率
```

```python
# y_test = y_test.ravel()
precision_score(y_test_pred, y_test)
```

### 调整参数

```python
datasets_array = datasets.values
np.random.shuffle(datasets_array) # 打乱顺序
x_tree_rfc, y_tree_rfc = datasets_array[: , :13], datasets_array[: , 13:]
y_tree_rfc = y_tree_rfc.ravel()
rfc_tree_1 = RandomForestClassifier(n_estimators=100, oob_score=True)
rfc_tree_1.fit(x_tree_rfc, y_tree_rfc)
```

```python
rfc_tree_1.oob_score_
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
# pipe = Pipeline([("scaler", StandardScaler()), ("rfc", RandomForestClassifier(n_estimators=100, oob_score=True))])
scaler = StandardScaler() # 标准化特征
rfc = RandomForestClassifier(n_estimators=100, oob_score=True)

x_tree_rfc_scaler = scaler.fit_transform(x_tree_rfc)

rfc.fit(x_tree_rfc_scaler, y_tree_rfc)
```

```python
rfc.oob_score_
```

## AdaBoosting 分类器

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
abc = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=4, max_features=5, min_samples_leaf=2), n_estimators=1000, learning_rate=0.8)  # 深度有很大影响,结果不稳定
abc.fit(x_train, y_train)
y_test_abc = abc.predict(x_test)
precision_score(y_test_abc, y_test)
```

```python
x_train_abc_scaler = scaler.fit_transform(x_train)  # 标准化后结果依旧不稳定
abc.fit(x_train_abc_scaler, y_train)
y_test_abc = abc.predict(x_test)
precision_score(y_test_abc, y_test)
```

## 线性分类器

```python
from sklearn.linear_model import SGDClassifier
```

```python
sgd = SGDClassifier()
sgd.fit(x_train, y_train)
```

```python
sgd.coef_, sgd.intercept_
```

```python
y_test_sgd = sgd.predict(x_test)
precision_score(y_test, y_test_sgd)
```

### 进行特征标准化

```python
pipe = Pipeline([('scaler', StandardScaler()), ('sgd', SGDClassifier())])
pipe.fit(x_train, y_train)
y_test_scaler_sgd = pipe.predict(x_test)
precision_score(y_test, y_test_scaler_sgd)
```
